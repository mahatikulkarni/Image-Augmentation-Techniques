OTSU

```python
import cv2
import os
```

```python
path = '/content/drive/MyDrive/Mid Range Cars/Normal/fiat/' # alter this as required
```

```python
def otsu(path):
    i=0
    for file in os.listdir(path):
        i=i+1
        img = cv2.imread(path+file)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        ret1,Otsu = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        dst = '/content/drive/MyDrive/Mid Range Cars/Otsu/fiat/'+file # alter this as required
        ret = cv2.imwrite(dst,Otsu)
        print("Augmenting...")
```

```python
otsu(path)
```

Inverse Otsu

```python
import cv2
import os
```

```python
path = '/content/drive/MyDrive/Mid Range Cars/Normal/fiat/' # alter this as required
```

```python
def otsuInv(path):
    i=0
    for file in os.listdir(path):
        i=i+1
        img = cv2.imread(path+file)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        ret2,Otsu_inv = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
        dst = '/content/drive/MyDrive/Mid Range Cars/Inverse Otsu/fiat/'+file # alter this as required
        ret = cv2.imwrite(dst,Otsu_inv)
        print("Augmenting...",ret)
```

Gaussian Noise (Salt and Pepper)

```python
import numpy as np
import random
import cv2
```

```python
path = '/content/drive/MyDrive/Mid Range Cars/Normal/tata/' # alter this as required
```

```python
def salt_pepper(path):
    i=0
    for file in os.listdir(path):
        i=i+1
        img = cv2.imread(path+file)
        gauss = np.random.normal(0,1,img.size)
        gauss = gauss.reshape(img.shape[0],img.shape[1],img.shape[2]).astype('uint8')
        dst = '/content/drive/MyDrive/Mid Range Cars/Gaussian Noise/tata/'+file   #destination folder path
        img_gauss = cv2.add(img,gauss)
        ret = cv2.imwrite(dst,img_gauss)
        print("Augmenting...")
```

```python
salt_pepper(path)
```
